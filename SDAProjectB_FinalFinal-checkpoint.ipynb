# Uber vs Taxi in NYC

## Motivation

The motivation for this project is based on the constantly increasing market share for ride sharing services, such as Uber, in comparison to the traditional Yellow Cab Taxis in New York City (NYC). It is often argued, that Uber, to a certain extent, is not competing directly with the Yellow Cabs of NYC, as they are servicing a different set of customers, specifically those who live outside of the city center of Manhatten in NYC. In this project, we attempt to address to which degree this statement is true. Hopefully, this will provide nuance to the debate regarding Uber and Yellow Cabs within the boroughs of NYC.  

The data for Uber is found at the [github](https://github.com/fivethirtyeight/uber-tlc-foil-response) of FiveThirtyEight.com, where the article that originally caught our attention, was published. 

The data for the Yellow Cabs was found at [NYC Taxi & Limousine Commission](https://www1.nyc.gov/site/tlc/about/tlc-trip-record-data.page) webpage.

We chose the data sets for the given time periods to directly compare Uber and Yellow Cabs in the same time period.

## Basic stats and data analysis

The two data sets we are working with concerns Uber and Yellow Taxi pickups, respectively, covering the months April through September of 2014 (half a year). Both data sets are combined from six separate data sets, each covering one month of operation. From the full data sets we have extracted the date, time, latitude and longitude of the pickup. The full set of Uber data consists of 4.5 million pickups and the full data set of Taxi data consists of 80.3 million pickups. 

In this project we are interested in investigating how popular Uber is compared to the historic Yellow Cab in New York City (NYC). In particular, we are interested in investigating at which time of day and at which day of the week Uber is most in demand, relative to the Yellow Cabs. Furthermore, we are interested in whether the popularity of Uber at given times and weekdays depends on which district of NYC is considered.

To answer these questions, we need a data set that shows the amount of Uber and Taxi pickups in each district of NYC for every day of the week and every hour of the day. 

Firstly, we need to convert both data sets to the same format. Since we will only consider day of the week and hour of the day in our analysis, we convert the pickup time for Uber and Taxi to only these two variables in two columns.

Secondly, we need to convert the latitude and longitude of each pickup into the corresponding district of NYC. To achieve this, we load a .csv file containing a grid over NYC with a district corresponding to each gridpoint. This grid is then interpolated to find the district of each pickup for both the Uber and Taxi data set.

Lastly, we need to convert the Uber and Taxi data sets of individual pickups into a combined data set of total number of pickups in each combination of district (Manhattan, Queens, Brooklyn, Bronx, Staten Island), day of the week (Monday - Sunday) and hour of the day (0 - 23). We do this by initializing an Uber and a Taxi dictionary with all combinations of (district, dayOfWeek, hourOfDay) that is updated while looping over the corresponding data set (either Uber or Taxi). The dictionary is then converted to a DataFrame and saved to a .csv to create the visualisations in Observable.

After completing this process, a number of things about the data become immediately clear. We see, unsurprisingly, that there are far more pickups (both Uber and Taxi) in Manhatten compared to other district. 

## Genre 

To tell our story, we want to use an interactive map of NYC.

We prefer visualizations that allow the reader to engage with the view, but under an overarching narrative structure that ensures the message is presented in easily digestible units and in a clear order. Additionally, where the incalculable amount of data together with the articles in the Observable references discussion are merged into a valuable annotated map.


Our visualization is a constant visual platform, that maintains the same graphical layout across options given to the user. In this way, the map is in the center of attention and makes it easy to compare the NYC ditricts to each other. We also use an interactive slider, where the user can choose an hour of the day, and see the corresponding distribution of uber pickups on the map.

To highlight important aspects of our data set, we have used feature distinction, by letting the reader pick a time of the day and hour of the week to see how the these factors influence the ratio of Uber pickups. 

To guide the reader through our visualization, we implemented an animation (animation transitions), that goes through the hours of the day. This is a simple way, to guide the user, and provides an easy overview of how the time of the day influences the data.

In the narrative structure, we used a user-directed path, where the reader is free to explore the data in any order he/she prefers by picking some of the selection features; one day of the week and one hour of the day and see the corresponding data. To make our visualization interactive, we implemented hover higlighting on our map, where the reader can see the exact ratio in percentage for each disrict.

To get our message clearly across, we made use of captions and annotations on the map.

## Visualizations

Our visualization has multiple interacting parts. The first, and central, part is a color-coded map of NYC that is divided into districts. Each district has a color ranging from purple to yellow based on the ratio of Uber pickups to total pickups in the corresponding district. The reader can select an hour of the day and a day of the week and the map will show the ratio for the chosen time and day. It is also possible to press play, and the animation will start going through every hour of the day given a specific day of the week. When hovering on a district, the exact ratio in percentage is displayed. 

Next to the map is a bar chart that displays the number of Uber and Taxi pickups for each district for the selected hour and day. The count axis of the bar chart is logarithmic to allow to viewer to more easily discern the differences in number of pickups in the different districts.

Below the map and the bar chart is a line graph. The reader can select a district and the graph shows Uber and Taxi pickups as a function of time on the day. This allows the reader to explore at what times of day Uber is more (or less) popular compared to Taxis.

Load Grid file and functions

```python
blocks = pd.read_csv('census_block_loc.csv')
blocks['Tract'] = blocks.BlockCode // 10000000000
blocks.head()
```

```python
# Create 2d grid from census_block_loc.csv
def convert_to_2d(lats,lons,values):
    latmin = 40.48
    lonmin = -74.28
    latmax = 40.93
    lonmax = -73.65
    lon_vals = np.mgrid[lonmin:lonmax:200j]
    lat_vals = np.mgrid[latmin:latmax:200j]
    map_values = np.zeros([200,200],'l')
    dlat = lat_vals[1] - lat_vals[0]
    dlon = lon_vals[1] - lon_vals[0]
    for lat,lon,value in zip(lats,lons,values):
        lat_idx = int(np.rint((lat - latmin) / dlat))
        lon_idx = int(np.rint((lon-lonmin) / dlon ))        
        if not np.isnan(value):
            map_values[lon_idx,lat_idx] = value
    return lat_vals,lon_vals,map_values

blocks_tmp = blocks[blocks.County.isin(['Bronx','Kings','New York','Queens','Richmond'])]
map_lats, map_lons,map_tracts_nyc = convert_to_2d(blocks_tmp.Latitude,blocks_tmp.Longitude,blocks_tmp.Tract)
map_lats, map_lons,map_tracts = convert_to_2d(blocks.Latitude,blocks.Longitude,blocks.Tract)

# Get district corresponding to lat, long from the grid map_tracts
def get_tract(lat,lon):
    latmin = 40.48
    lonmin = -74.28
    latmax = 40.93
    lonmax = -73.65
    dlat = (latmax-latmin) / 199
    dlon = (lonmax-lonmin) / 199
    if (latmin<lat<latmax) and (lonmin<lon<lonmax):
        lat_idx = int(np.rint((lat - latmin) / dlat))
        lon_idx = int(np.rint((lon-lonmin) / dlon )) 
        return map_tracts[lon_idx,lat_idx]
    return 0

fips_codes = {36061:'Manhattan',36081:'Queens',36047:'Brooklyn',
              36005:'Bronx',36085:'Staten Island',36059:'Nassau',36119:'Westchester',
              34017:'NJ',34013:'NJ',34003:'NJ',34039:'NJ',
              34031:'NJ',34023:'NJ',34025:'NJ'#,0:'Unknown'
             }
districts = list(set(fips_codes.values()))
```

## Creating data set for visualisations

We create a data set with the number of both Uber and Taxi pickups in each of the five major districts of New York City. First we create an empty dictionary and fill it in by looping over the full dataset.

```python
UberDict = dict(((week, hour, district),0) for week in weekDaySet for hour in hourOfDaySet for district in districtSet)
          

for row in range(len(dataU)):
    key = (dataU['Weekday'].iloc[row], dataU['HourOfDay'].iloc[row], dataU['District'].iloc[row])
    UberDict[key] = UberDict[key] + 1

    
UberDict
    
```

We do the same for the Taxi data.

```python
TaxiDict = dict(((week, hour, district),0) for week in weekDaySet for hour in hourOfDaySet for district in districtSet) 
for row in range(len(dataY)):
    key = (dataY['Weekday'].iloc[row], dataY['HourOfDay'].iloc[row], dataY['District'].iloc[row])
    TaxiDict[key] = TaxiDict[key] + 1

    
TaxiDict
```

Create a dataframe with all the relevant data from the two dictionaries created above.

```python
districtSet = {x for x in districtSet if pd.notna(x)}
N = len(weekDaySet)*len(hourOfDaySet)*len(districtSet)
data = pd.DataFrame(index=range(N), columns=['UberPickups','TaxiPickups','Weekday', 'HourOfDay','District'])
row = 0
for week in weekDaySet:
    for hour in hourOfDaySet:
        for district in districtSet:
            key = (week, hour, district)
            data.iloc[row] = [UberDict[key], TaxiDict[key], week, hour, district]
            row = row + 1
data
```

Export the new data set to a .csv for import into Observable.

```python
data.to_csv("UberTaxiPickups.csv", sep = ',')
```
